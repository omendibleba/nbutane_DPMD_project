```python
# Put everything in a function
def rename_atom_types(data_file, new_data_file):
    # Read in the data file
    data = np.genfromtxt(data_file, dtype=str, skip_header=9)
    types = data[:,1]
    # loop over types and replace 2 by 1 and 3 by 2
    types_new = np.zeros(len(types), dtype=int)
    for i in range(len(types)):
        if types[i] == '2':
            types_new[i] = '1'
        elif types[i] == '3':
            types_new[i] = '2'
        elif types[i] == '1':
            types_new[i] = '1'
    # substitute the old types with the new types in the data array
    data[:,1] = types_new
    # write the new data file using the new data array
    np.savetxt(new_data_file, data, fmt='%s', header='LAMMPS data file via VMD', comments='')
    return
```

```python
# Define function that reads in the data file and deletes the last three columns
def delete_last_three_columns(data_file,new_data_file):
    # Read in the data file
    data = np.genfromtxt(data_file, dtype=str, skip_header=9)
    # Delete the last three columns
    data = np.delete(data, np.s_[-3:], axis=1)
    # write the new data file using the new data array
    np.savetxt(new_data_file, data, fmt='%s', header='LAMMPS data file via VMD', comments='')
    return data
```

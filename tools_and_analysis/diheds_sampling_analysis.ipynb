```python
# NOtebook that read the sapmled dihedrals calculated with VMD and plot a histogram.
# first use the tcl script to read an xyz trajectory and calculate the dihedral angle for each frame given the atoms indexes 

# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Lets read first the dihedral angles from the AIMD simulations 

# Read dihedral sampling analaysis obtained with VMD. for AIMD trajectory 

aimd_diheds_6but_272K_20ps_molec_1 = np.loadtxt('../aimd/T_272/diheds_sampled_aimd_272/dihedrals_aimd_272K_molec_1.txt')
aimd_diheds_6but_272K_20ps_molec_2 = np.loadtxt('../aimd/T_272/diheds_sampled_aimd_272/dihedrals_aimd_272K_molec_2.txt')
aimd_diheds_6but_272K_20ps_molec_3 = np.loadtxt('../aimd/T_272/diheds_sampled_aimd_272/dihedrals_aimd_272K_molec_3.txt')
aimd_diheds_6but_272K_20ps_molec_4 = np.loadtxt('../aimd/T_272/diheds_sampled_aimd_272/dihedrals_aimd_272K_molec_4.txt')
aimd_diheds_6but_272K_20ps_molec_5 = np.loadtxt('../aimd/T_272/diheds_sampled_aimd_272/dihedrals_aimd_272K_molec_5.txt')
aimd_diheds_6but_272K_20ps_molec_6 = np.loadtxt('../aimd/T_272/diheds_sampled_aimd_272/dihedrals_aimd_272K_molec_6.txt')
```

```python
## Determine lentght of the txt file, which should be the same for all 

length = len(aimd_diheds_6but_272K_20ps_molec_1)

# Define the number of molecules which dihedral sare to be sampled 
n_molecs = 6 

# Define the total number of dihedrals to be sampled
total_diheds = length * n_molecs
total_diheds
```

```python
# Create array of zeros to store all th esampled diheds 
all_diheds = np.zeros((total_diheds))

for i in range(len(aimd_diheds_6but_272K_20ps_molec_1)):
    all_diheds[i] = aimd_diheds_6but_272K_20ps_molec_1[i]
    all_diheds[i + length] = aimd_diheds_6but_272K_20ps_molec_2[i]
    all_diheds[i + 2*length] = aimd_diheds_6but_272K_20ps_molec_3[i]
    all_diheds[i + 3*length] = aimd_diheds_6but_272K_20ps_molec_4[i]
    all_diheds[i + 4*length] = aimd_diheds_6but_272K_20ps_molec_5[i]
    all_diheds[i + 5*length] = aimd_diheds_6but_272K_20ps_molec_6[i]
```

```python
plt.figure(figsize=(10, 8), facecolor='white')
plt.hist((all_diheds),bins=30, alpha=0.7,)
plt.title('Sampled dihedral angles in AIMD at 272 K',fontsize=22,fontweight='bold')
plt.xlabel('Dihedral angle (degrees)',fontsize=18,fontweight='bold')
plt.ylabel('Frequency',fontsize=18,fontweight='bold')
plt.xticks(fontsize=18)
# Define the limits and ticks of x axis from -180 to 180 degrees
plt.xlim(-180,180)
plt.xticks(np.arange(-180, 181, 30))
plt.yticks([],fontsize=14)
#plt.legend(fontsize=11,loc='upper center')
plt.show()
```

```python
## Now for the AIMD Data at 272 K

# read trajectory data for CLMD

clmd_diheds_168but_272K_1ns_molec_1 = np.loadtxt('../classical_MD/T_272K/diheds_sampled_clmd_272/dihed_clmd_but1_168molec_272K_1ns_mol_1.txt')
clmd_diheds_168but_272K_1ns_molec_2 = np.loadtxt('../classical_MD/T_272K/diheds_sampled_clmd_272/dihed_clmd_but1_168molec_272K_1ns_mol_2.txt')
clmd_diheds_168but_272K_1ns_molec_3 = np.loadtxt('../classical_MD/T_272K/diheds_sampled_clmd_272/dihed_clmd_but1_168molec_272K_1ns_mol_3.txt')
clmd_diheds_168but_272K_1ns_molec_4 = np.loadtxt('../classical_MD/T_272K/diheds_sampled_clmd_272/dihed_clmd_but1_168molec_272K_1ns_mol_4.txt')
clmd_diheds_168but_272K_1ns_molec_5 = np.loadtxt('../classical_MD/T_272K/diheds_sampled_clmd_272/dihed_clmd_but1_168molec_272K_1ns_mol_5.txt')
clmd_diheds_168but_272K_1ns_molec_6 = np.loadtxt('../classical_MD/T_272K/diheds_sampled_clmd_272/dihed_clmd_but1_168molec_272K_1ns_mol_6.txt')
```

```python
## Determine lentght of the txt file, which should be the same for all 

length = len(clmd_diheds_168but_272K_1ns_molec_1)
length

# Define the number of molecules which dihedral sare to be sampled 
n_molecs = 6 

# Define the total number of dihedrals to be sampled
total_diheds = length * n_molecs
total_diheds
```

```python
# Create aux array to store the dihedral data from each file 

total_diheds_sampled = np.zeros((total_diheds))

for i in range(len(clmd_diheds_168but_272K_1ns_molec_1)):
    total_diheds_sampled[i] = clmd_diheds_168but_272K_1ns_molec_1[i]

for i in range(len(clmd_diheds_168but_272K_1ns_molec_2)):
    total_diheds_sampled[i+length] = clmd_diheds_168but_272K_1ns_molec_2[i]

for i in range(len(clmd_diheds_168but_272K_1ns_molec_3)):
    total_diheds_sampled[i+length*2] = clmd_diheds_168but_272K_1ns_molec_3[i]

for i in range(len(clmd_diheds_168but_272K_1ns_molec_4)):
    total_diheds_sampled[i+length*3] = clmd_diheds_168but_272K_1ns_molec_4[i]

for i in range(len(clmd_diheds_168but_272K_1ns_molec_5)):
    total_diheds_sampled[i+length*4] = clmd_diheds_168but_272K_1ns_molec_5[i]

for i in range(len(clmd_diheds_168but_272K_1ns_molec_6)):
    total_diheds_sampled[i+length*5] = clmd_diheds_168but_272K_1ns_molec_6[i]
```

```python
# Plot histogram of dihedral angles for CLMD

plt.hist((clmd_diheds_168but_272K_1ns_molec_1,clmd_diheds_168but_272K_1ns_molec_2,clmd_diheds_168but_272K_1ns_molec_3,
clmd_diheds_168but_272K_1ns_molec_4,clmd_diheds_168but_272K_1ns_molec_5,clmd_diheds_168but_272K_1ns_molec_6), bins=20)
plt.title('CLMD dihedral angles for 168 butane molecules at 272K')
plt.xlabel('Dihedral angle (degrees)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 8), facecolor='white')
plt.hist((total_diheds_sampled),bins=30, alpha=0.7,)
plt.title('Sampled dihedral angles in CLMD' ,fontsize=22,fontweight='bold')
plt.xlabel('Dihedral angle (degrees)',fontsize=18,fontweight='bold')
plt.ylabel('Frequency',fontsize=18,fontweight='bold')
plt.xticks(fontsize=18)
# Define the limits and ticks of x axis from -180 to 180 degrees
plt.xlim(-180,180)
plt.xticks(np.arange(-180, 181, 30))
plt.yticks([],fontsize=18)
#plt.legend(fontsize=11,loc='upper center')
plt.show()
```

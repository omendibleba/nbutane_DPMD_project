```python
""" NOtee book to vizualize AIMD results and confirm equilibration.

The energy files loaded in this notebook are obatained from thhe simulation in the aimd directory.

 """


import numpy as np
import matplotlib.pyplot as plt
```

```python
# define path to energy file of AIMD sim at 160 K 

aimd_160K = "../aimd/T_160K/Butane_solv_250K-1.ener"
```

```python
# Define function to calculate averages and plot temperature and potential energy of the AIMD simulation 

def plot_aimd_equil(aimd_file):

    file = np.loadtxt(aimd_file)

    # Determine average temperature from the file 

    temp_ave_272 = np.mean(file[:,3])
    temp_ave_272
    max(file[:,1]/1000) # column 1 is time in fs(divided by 1000 to get fs), column 3 is temperature in K

    # Set figure size and background color
    plt.figure(figsize=(9,6), facecolor='white')

    # Plot the temperature vs time. Columns 4 vs 2
    plt.plot(file[:,1]/1000,file[:,3])
    plt.title("AIMD Temperature eqilibration ", fontsize=18, fontweight='bold')

    # Plot horizontal line at the average temperature using the same x valeus as the plot above
    # plt.axhline(y=temp_ave_272, xmin=0, xmax=1000, linestyle='--')
    plt.axhline(y=temp_ave_272,color='r', linestyle='-',label='Average Temperature = {} K'.format(round(temp_ave_272,2)))


    plt.xlabel("Time (ps)",fontsize=16,fontweight="bold")
    plt.ylabel("Temperature (K)",fontsize=16,fontweight="bold")
    plt.xlim(min(file[:,1]/1000),max(file[:,1]/1000))
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.legend(loc='lower right',fontsize=12)
    plt.show()

    # Determine the average energy from the file

    pot_ener_ave_272 = np.mean(file[:,4])

    # DO the same plot for the potential energy 
    plt.figure(figsize=(9,6), facecolor='white')
    plt.plot(file[:,1]/1000,file[:,4])
    plt.title("AIMD Potential Energy eqilibration ", fontsize=18, fontweight='bold')

    # Plot horizontal line at the average energy using the same x valeus as the plot above
    plt.axhline(y=pot_ener_ave_272, linestyle='-',color='red',label='Average Energy = {} Hartree'.format(round(pot_ener_ave_272,2)))
    plt.xlabel("Time (ps)",fontsize=16,fontweight="bold")
    plt.ylabel("Potential Energy (Hartree)",fontsize=16,fontweight="bold")
    plt.xlim(min(file[:,1]/1000),max(file[:,1]/1000))
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.legend(loc='lower right',fontsize=12)
    plt.show()
```

```python
# Plot results fro AIMD at 160 K

plot_aimd_equil(aimd_160K)
```

```python
# Define path for AIMD at 200 K

aimd_200K = "../aimd/T_200/Butane_solv_250K-1.ener"

# Plot results for AIMD at 200 K

plot_aimd_equil(aimd_200K)
```

```python
# Define path for AIMD at 250 K

aimd_250K = "../aimd/T_250/Butane_solv_250K-1.ener"

# Plot results for AIMD at 250 K

plot_aimd_equil(aimd_250K)
```

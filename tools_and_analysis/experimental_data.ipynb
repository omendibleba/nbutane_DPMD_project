```python
# Define a function that takes experimental and actual values list and returns the absolute error percent.
def abs_error(exp,act):
    error = []
    for i in range(len(exp)):
        error.append(100*abs(exp[i]-act[i])/exp[i])
    return error
    
```

```python
# Use function to determine the error between liq_dens_exp and liq_dens_md
abs_error(liq_dens_exp,liq_dens_md)
```

```python
# Determine the error between gas_dens_exp and dens_md
abs_error(gas_dens_used,dens_md)
```

```python
import numpy as np 
import matplotlib.pyplot as plt
import lammps_logfile as lmplog
```

```python
## STrat with liquid simulation at T= 160 K
# Read in the log file
log = lmplog.File('../classical_MD/T_160K/nvt_production_160K_1ns/log.lammps')
```

```python
## Get the time, temperature, total energy, pressure and enthalpy
time = log.get('Time')
temp = log.get('Temp')
toten = log.get('TotEng')
enthalpy = log.get('Enthalpy')
pressure = log.get('Press')
```

```python
## plot four sublplots in one figure, 2 rows and two columns. plot the temperature, total energy, pressure and enthalpy as a function of time
fig, ax = plt.subplots(2,2, figsize=(10,10))

# Determine average temperature amd plot as a horizontal line.
avg_temp = np.mean(temp)

ax[0,0].plot(time, temp,label = 'Temperature = {} K'.format(round(avg_temp,2)))
ax[0,0].axhline(avg_temp, color='r', linestyle='--')
ax[0,0].set_xlabel('Time (ps)', fontsize=14, fontweight='bold')
# Increase label size and bold it 
ax[0,0].set_ylabel('Temperature (K)', fontsize=14, fontweight='bold')
ax[0,0].set_title('Temperature', fontsize=16, fontweight='bold')
ax[0,0].legend()


# Determine average total energy amd plot as a horizontal line
avg_toten = np.mean(toten)

ax[0,1].plot(time, toten)
ax[0,1].axhline(avg_toten, color='r', linestyle='--', label = 'Average Total Energy = {} kcal/mol'.format(round(avg_toten,3)))

ax[0,1].set_xlabel('Time (ps)', fontsize=14, fontweight='bold')
ax[0,1].set_ylabel('Total Energy (kcal/mol)', fontsize=14, fontweight='bold')
ax[0,1].set_title('Total Energy', fontsize=16, fontweight='bold')
ax[0,1].legend()


# Determine average pressure amd plot as a horizontal line
avg_pressure = np.mean(pressure)

ax[1,0].plot(time, pressure)
ax[1,0].axhline(avg_pressure, color='r', linestyle='--', label = 'Average Pressure = {} atm'.format(round(avg_pressure,2)))
ax[1,0].set_xlabel('Time (ps)', fontsize=14, fontweight='bold')
ax[1,0].set_ylabel('Pressure (atm)', fontsize=14, fontweight='bold')
ax[1,0].set_title('Pressure', fontsize=16, fontweight='bold')
ax[1,0].legend()


# Determine average enthalpy amd plot as a horizontal line
avg_enthalpy = np.mean(enthalpy)

ax[1,1].plot(time, enthalpy)
ax[1,1].axhline(avg_enthalpy, color='r', linestyle='--', label = 'Average Enthalpy = {} kcal/mol'.format(round(avg_enthalpy,3)))
ax[1,1].set_xlabel('Time (ps)', fontsize=14, fontweight='bold')
ax[1,1].set_ylabel('Enthalpy (kcal/mol)', fontsize=14, fontweight='bold')
ax[1,1].set_title('Enthalpy', fontsize=16, fontweight='bold')
ax[1,1].legend()

plt.tight_layout()

plt.show()
```

```python
## Define a function to make the plot from above

def plot_enthalpy(logfile):
    # Read in the log file
    log = lmplog.File(logfile)

    ## Get the time, temperature, total energy, pressure and enthalpy
    time = log.get('Time') /1000
    temp = log.get('Temp')
    toten = log.get('TotEng')
    enthalpy = log.get('Enthalpy')
    pressure = log.get('Press')

    ## plot four sublplots in one figure, 2 rows and two columns. plot the temperature, total energy, pressure and enthalpy as a function of time
    fig, ax = plt.subplots(2,2, figsize=(10,10))

    # Determine average temperature amd plot as a horizontal line.
    avg_temp = np.mean(temp)

    ax[0,0].plot(time, temp,label = 'Temperature = {} K'.format(round(avg_temp,2)))
    ax[0,0].axhline(avg_temp, color='r', linestyle='--')
    ax[0,0].set_xlabel('Time (ps)', fontsize=14, fontweight='bold')
    # Increase label size and bold it 
    ax[0,0].set_ylabel('Temperature (K)', fontsize=14, fontweight='bold')
    ax[0,0].set_title('Temperature', fontsize=16, fontweight='bold')
    # Increase thick size
    ax[0,0].tick_params(axis='both', which='major', labelsize=12)
    ax[0,0].legend(fontsize=11)


    # Determine average total energy amd plot as a horizontal line
    avg_toten = np.mean(toten)

    ax[0,1].plot(time, toten)
    ax[0,1].axhline(avg_toten, color='r', linestyle='--', label = 'Average Total Energy = {} kcal/mol'.format(round(avg_toten,2)))

    ax[0,1].set_xlabel('Time (ps)', fontsize=14, fontweight='bold')
    ax[0,1].set_ylabel('Total Energy (kcal/mol)', fontsize=14, fontweight='bold')
    ax[0,1].set_title('Total Energy', fontsize=16, fontweight='bold')
    # Increase thick size
    ax[0,1].tick_params(axis='both', which='major', labelsize=12)
    ax[0,1].legend(fontsize=11)


    # Determine average pressure amd plot as a horizontal line
    avg_pressure = np.mean(pressure)

    ax[1,0].plot(time, pressure)
    ax[1,0].axhline(avg_pressure, color='r', linestyle='--', label = 'Average Pressure = {} atm'.format(round(avg_pressure,2)))
    ax[1,0].set_xlabel('Time (ps)', fontsize=14, fontweight='bold')
    ax[1,0].set_ylabel('Pressure (atm)', fontsize=14, fontweight='bold')
    ax[1,0].set_title('Pressure', fontsize=16, fontweight='bold')
    # Increase thick size
    ax[1,0].tick_params(axis='both', which='major', labelsize=12)
    ax[1,0].legend(fontsize=11)


    # Determine average enthalpy amd plot as a horizontal line
    avg_enthalpy = np.mean(enthalpy)

    ax[1,1].plot(time, enthalpy)
    ax[1,1].axhline(avg_enthalpy, color='r', linestyle='--', label = 'Average Enthalpy = {} kcal/mol'.format(round(avg_enthalpy,2)))
    ax[1,1].set_xlabel('Time (ps)', fontsize=14, fontweight='bold')
    ax[1,1].set_ylabel('Enthalpy (kcal/mol)', fontsize=14, fontweight='bold')
    ax[1,1].set_title('Enthalpy', fontsize=16, fontweight='bold')
    # Increase thick size
    ax[1,1].tick_params(axis='both', which='major', labelsize=12)
    ax[1,1].legend(fontsize=11)

    plt.tight_layout()

    plt.show()

    # Print average ehthalpy
    # Convert to kJ/mol 
    print('Average Enthalpy = {} kJ/mol'.format(round(avg_enthalpy*4.184,2)))
    
    print('')

    return avg_enthalpy



    
```

```python
log_liq_160 = '../classical_MD/T_160K/nvt_production_160K_1ns/log.lammps'

h_ave_160 = plot_enthalpy(log_liq_160)
```

```python
log_liq_175 = '../classical_MD/T_160K/nvt_production_175K_1ns/log.lammps'

h_ave_175 = plot_enthalpy(log_liq_175)
```

```python
# Define a function to to determine heat capccacity (Cv) using the enthalpies 

def heat_capacity(h_T1, h_T2, T1, T2):
    # Calculate the heat capacity (Cv) using the enthalpies
    Cv_kcal = (h_T2 - h_T1) / (T2 - T1)
    
    # Convert to kJ/mol
    Cv_kJ = Cv_kcal * 4.184
    
    # Print the heat capacity
    print('Heat Capacity (Cv) = {} kcal/mol/K'.format(round(Cv_kcal,2)))
    print('Heat Capacity (Cv) = {} kJ/mol/K'.format(round(Cv_kJ,2)))
    print('')
    return Cv_kJ

```

```python
## Heat capacity (Cv) for liquid butane between 160 K and 175 K

Cv_liq_160 = heat_capacity(h_ave_175, h_ave_160, 175, 160)
```

```python
# Define Cv from literature at 160 K 
Cv_exp_160 = 87.36 # J/mol K
```

```python
# Define a function to calculate the percent absolute error between the calculated and experimental Cv
def percent_error(Cv_calc, Cv_exp):
    # Calculate the percent error
    percent_error = abs((Cv_calc - Cv_exp) / Cv_exp) * 100
    
    # Print the percent error
    print('Percent Error = {} %'.format(round(percent_error,2)))
    print('')
    return percent_error
```

```python
Cv_error_160 = percent_error(Cv_liq_160, Cv_exp_160)
```

```python
# THis noteboplk read andd plot the reusltd in the logfile of LAMMPS simulations from the NVT simulation of 1 ns

import numpy as np
import matplotlib.pyplot as plt
import lammps_logfile
```

```python
def nvt_analysis_1ns(log_file_name,P,T):

    
    #read log file
    log = lammps_logfile.File(log_file_name)
    step = log.get("Step") ## Read steps
    time = log.get("Time")
    # Change the figure size
    plt.figure(figsize=[12, 12])
    
    # Set title 
    plt.suptitle('NVT Production run Equilibrium Analysis P(atm)= '+ str(P)+' T(K)= '+str(T), y=1.12, fontsize=19, fontweight='bold')
    plt.subplots_adjust(left=0.13, right=0.93, top=1.0, bottom= 0.27, wspace= 0.3, hspace=0.3)
    
    # Plot the subplots
    # Plot 1:  Temperature
    plt.subplot(2, 2, 1)
    temp = log.get("Temp")

    # Temperature average
    temp_ave = round(np.average(temp[500:]),2)
    plt.plot(time,temp,color="blue")
    plt.hlines(temp_ave,xmin=min(time[500:]),xmax=max(time),color="red",label = " Temperature = {} K".format(temp_ave))
    #plt.xlabel("Step")
    plt.xlabel("Time (ps)",fontsize=16 , fontweight='bold')
    plt.ylabel("Temperature (K)",fontsize=16 , fontweight='bold')
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.legend(fontsize=12)

    # Plot the subplots
    # Plot 2:  Presure
    plt.subplot(2, 2, 2)
    press = log.get("Press")
    press_ave = round(np.average(press[500:]),4)
    plt.plot(time,press,color="blue")
    plt.hlines(press_ave,xmin=min(time[500:]),xmax=max(time),color="red",label = "Ave Pressure = {} atm".format(press_ave))
    plt.xlabel("Time (ps)",fontsize=16 , fontweight='bold')
    plt.ylabel("Pressure (atm)",fontsize=16 , fontweight='bold')
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.legend(fontsize=12)

    # Plot the subplots
    # Plot 3:  Potential Energy 
    plt.subplot(2, 2, 3)
    potEng = log.get("PotEng")
    potEng_ave = round(np.average(potEng[500:]),3)
    plt.plot(time,potEng,color="blue")
    plt.hlines(potEng_ave,xmin=min(time[500:]),xmax=max(time),color="red",label = "Ave Potential Energy = {} Kcal/mol".format(potEng_ave))
    #plt.title("Potential Energy ")
    plt.xlabel("Time (ps)",fontsize=16 , fontweight='bold')
    plt.ylabel("Potential Energy (Kcal/mol)",fontsize=16 , fontweight='bold')
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.legend(fontsize=12,loc='lower right')

 # Plot the subplots
    # Plot 4:  Density
    plt.subplot(2, 2, 4)
    dens = log.get("Density")
    dens_ave = round(np.average(dens[500:]),6)
    #plt.plot(step,dens,color="blue",label = "Average Density = {}  ".format(dens_ave))
    plt.plot(time,dens,color="blue",label = "Average Density = {} g/$cm^3$ ".format(dens_ave))
    #plt.hlines(dens_ave,xmin=min(time[500:]),xmax=max(time),color="red",label = "Average Density = {} g/$cm^3$ ".format(dens_ave))
    plt.xlabel("Time (ps)",fontsize=16 , fontweight='bold')
    plt.ylabel("Density (g/cm^3)",fontsize=16 , fontweight='bold')
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.legend(fontsize=12)

    print(" Average over last 20ps of simulation")
    print('Temperature: ',round(np.average(temp[1800:]),3),' K')
    print('Presure: ',round(np.average(press[1800:]),3),' atm')
    print('Potencial Energy: ',round(np.average(temp[1800:]),3),' Kcal/mol')
    print('Density: ',round(np.average(dens[1800:]),5),' g/cm3')

    plt.show()
```

```python
## Use function to read and plot log file of the NVT equilibration of butane at 160 K 

liq_nvt_T_160_file = '../classical_MD/T_160K/nvt_production_160K_1ns/log.lammps'

nvt_analysis_1ns(liq_nvt_T_160_file,1,160)
```
